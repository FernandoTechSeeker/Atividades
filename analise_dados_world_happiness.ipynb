## 1Ô∏è‚É£C√°lculo da Tabuada
 üìå Objetivo: Criar uma fun√ß√£o que exiba a tabuada de um n√∫mero de 1 a 10.

üìå Passos:
- ‚úî Criar uma fun√ß√£o calcular_tabuada(numero: int).
- ‚úî Validar se o n√∫mero est√° no intervalo de 1 a 10.
- ‚úî Exibir a tabuada do n√∫mero informado.
- ‚úî Se o n√∫mero for inv√°lido, exibir uma mensagem de erro.
- ‚úî Melhorias sugeridas: Criar um loop para novas tentativas e melhorar a formata√ß√£o da sa√≠da.


```python
# Fun√ß√£o para calcular e exibir a tabuada com tipagem expl√≠cita
def calcular_tabuada(numero: int) -> None:
    """
    Fun√ß√£o que exibe a tabuada de um n√∫mero entre 1 e 10.

    Par√¢metros:
    numero (int): N√∫mero para calcular a tabuada.

    Retorno:
    Exibe a tabuada do n√∫mero escolhido ou uma mensagem de erro se o n√∫mero estiver fora do intervalo.
    """
    print("\n" + "=" * 25)
    print(f"üìå Tabuada do {numero}".center(25))
    print("=" * 25)

    for i in range(1, 11):
        print(f"{numero} x {i:2} = {numero * i:3}")  # Formata√ß√£o para alinhar os resultados

    print("=" * 25, "\n")

# Loop para permitir novas tentativas caso o usu√°rio insira um valor inv√°lido
while True:
    try:
        num = int(input("üîπ Digite um n√∫mero entre 1 e 10 para ver a tabuada (ou 0 para sair): "))

        if num == 0:
            print("‚úÖ Saindo do programa. At√© mais!")
            break  # Encerra o loop se o usu√°rio digitar 0

        if 1 <= num <= 10:
            calcular_tabuada(num)  # Chamada da fun√ß√£o para exibir a tabuada
        else:
            print("‚ùå Erro: O n√∫mero deve estar entre 1 e 10. Tente novamente.")

    except ValueError:
        print("‚ùå Erro: Por favor, insira um n√∫mero inteiro v√°lido.")
```

##2Ô∏è‚É£ Escolha e An√°lise de um Dataset P√∫blico

Dataset Escolhido: **World Happiness Report 2022**

- ‚úî Descri√ß√£o: Dados sobre felicidade global, incluindo PIB per capita, suporte social e expectativa de vida.
- ‚úî Analise estat√≠sticas dos pa√≠ses e compara√ß√µes entre anos.
- ‚úî Visualizar informa√ß√µes iniciais com .info() e .describe().
- ‚úî Selecionar colunas relevantes e discutir os dados.

```python
# üìå Importando a biblioteca Pandas para manipula√ß√£o de dados
import pandas as pd

# üìå Carregando o dataset de Felicidade Mundial 2022
# Este dataset cont√©m informa√ß√µes sobre diversos pa√≠ses e seus √≠ndices de felicidade.
df_happiness = pd.read_csv("World_Happiness_2022.csv")

# üìå Exibindo as primeiras linhas do dataset para entender sua estrutura
# Isso nos ajuda a visualizar como os dados est√£o organizados.
print("üîπ Primeiras 5 linhas do dataset:")
display(df_happiness.head())

# üìå Exibindo informa√ß√µes gerais do dataset
# Aqui verificamos quantas colunas temos, o tipo de dados e se h√° valores nulos.
print("\nüîπ Informa√ß√µes gerais do dataset:")
df_happiness.info()

# üìå Exibindo estat√≠sticas descritivas
# Estat√≠sticas como m√©dia, desvio padr√£o, valores m√≠nimos e m√°ximos s√£o √∫teis para entender os dados.
print("\nüîπ Estat√≠sticas gerais do dataset:")
display(df_happiness.describe())

# üìå Selecionando colunas mais relevantes para an√°lise
# Como o dataset pode ter muitas colunas, escolhemos as mais importantes para esta an√°lise.
colunas_relevantes = ['Country', 'Happiness score', 'Explained by: GDP per capita',
                      'Explained by: Social support', 'Explained by: Healthy life expectancy']

# üìå Filtrando o DataFrame com as colunas corretas
# Aqui criamos um novo DataFrame contendo apenas as colunas selecionadas.
df_happiness_filtrado = df_happiness[colunas_relevantes]

# üìå Exibindo dataset filtrado
# Agora, mostramos as primeiras linhas desse novo DataFrame para verificar se a sele√ß√£o foi feita corretamente.
print("üîπ Dataset filtrado com colunas relevantes:")
display(df_happiness_filtrado.head())
```

##3Ô∏è‚É£ Expans√£o da An√°lise com o World Happiness Report
Nesta etapa, daremos continuidade √† an√°lise do Relat√≥rio Mundial da Felicidade, incorporando os dados do ano de 2020 ao conjunto j√° utilizado (2019 e 2022). O objetivo √© unificar essas informa√ß√µes para identificar padr√µes e varia√ß√µes na felicidade global ao longo do tempo.

üîπ Objetivos desta etapa:
- ‚úî Carregar os dados do ano de 2020 e verificar sua estrutura.
- ‚úî Tratar e limpar os dados, garantindo que estejam padronizados para jun√ß√£o.
- ‚úî Unificar os datasets de 2019, 2020 e 2022 para facilitar a an√°lise comparativa.
- ‚úî Utilizar os m√©todos do Pandas (merge, concat) para consolidar os dados corretamente.
- ‚úî Exibir os dados finais e estat√≠sticas para verificar padr√µes e tend√™ncias.

```python
# üìå Importando a biblioteca Pandas
import pandas as pd

# üìå Carregando os datasets de Felicidade Mundial para os anos de 2019, 2020 e 2022
df_2019 = pd.read_csv("World_Happiness_2019.csv")
df_2020 = pd.read_csv("World_Happiness_2020.csv")
df_2022 = pd.read_csv("World_Happiness_2022.csv")

# üìå Exibir as primeiras linhas de cada dataset para entender sua estrutura
print("üîπ Primeiras linhas do dataset de 2019:")
display(df_2019.head())

print("\nüîπ Primeiras linhas do dataset de 2020:")
display(df_2020.head())

print("\nüîπ Primeiras linhas do dataset de 2022:")
display(df_2022.head())

# üìå Exibir informa√ß√µes gerais dos datasets para identificar colunas e valores nulos
print("üîπ Informa√ß√µes do Dataset 2019:")
df_2019.info()

print("\nüîπ Informa√ß√µes do Dataset 2020:")
df_2020.info()

print("\nüîπ Informa√ß√µes do Dataset 2022:")
df_2022.info()

# üìå Selecionando colunas mais relevantes para an√°lise
# Como os datasets possuem colunas diferentes, ajustamos os nomes conforme necess√°rio
colunas_relevantes_2019 = ['Country or region', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy']
colunas_relevantes_2020 = ['Country name', 'Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy']
colunas_relevantes_2022 = ['Country', 'Happiness score', 'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy']

# üìå Renomeando colunas para manter um padr√£o √∫nico antes da uni√£o dos datasets
df_2019 = df_2019[colunas_relevantes_2019].rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'GDP per capita',
    'Social support': 'Social Support',
    'Healthy life expectancy': 'Healthy Life Expectancy'
})

df_2020 = df_2020[colunas_relevantes_2020].rename(columns={
    'Country name': 'Country',
    'Ladder score': 'Happiness Score',
    'Logged GDP per capita': 'GDP per capita',
    'Social support': 'Social Support',
    'Healthy life expectancy': 'Healthy Life Expectancy'
})

df_2022 = df_2022[colunas_relevantes_2022].rename(columns={
    'Country': 'Country',
    'Happiness score': 'Happiness Score',
    'Explained by: GDP per capita': 'GDP per capita',
    'Explained by: Social support': 'Social Support',
    'Explained by: Healthy life expectancy': 'Healthy Life Expectancy'
})

# üìå Adicionando a informa√ß√£o do ano a cada dataset
df_2019['Year'] = 2019
df_2020['Year'] = 2020
df_2022['Year'] = 2022

# üìå Unindo os datasets para an√°lise completa
df_completo = pd.concat([df_2019, df_2020, df_2022], ignore_index=True)

# üìå Exibir os primeiros dados combinados
print("üîπ Dados combinados ap√≥s a unifica√ß√£o:")
display(df_completo.head(1000))

# üìå Exibir estat√≠sticas descritivas dos dados combinados
print("\nüîπ Estat√≠sticas dos Dados Combinados:")
display(df_completo.describe())

# üìå Exibir a quantidade total de pa√≠ses analisados
print("\nüîπ Quantidade de pa√≠ses analisados:", df_completo['Country'].nunique())
```

##4Ô∏è‚É£ Visualiza√ß√£o de Dados ‚Äì Gr√°ficos
Nesta se√ß√£o, iremos criar e interpretar gr√°ficos baseados nos dados do World Happiness Report para identificar padr√µes e insights relevantes. Utilizaremos as bibliotecas Matplotlib e Seaborn para gerar as visualiza√ß√µes.

Objetivos desta etapa:
- ‚úîCriar tr√™s gr√°ficos para representar os dados
- ‚úî Identificar padr√µes e tend√™ncias atrav√©s da visualiza√ß√£o
- ‚úî Explicar os insights obtidos a partir dos gr√°ficos



```python
# üìå Importando bibliotecas essenciais para an√°lise e visualiza√ß√£o de dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configura√ß√£o do estilo dos gr√°ficos para uma melhor visualiza√ß√£o
sns.set_style("whitegrid")

```

Verificamos anteriormente que a unifica√ß√£o dos datasets funcionou corretamente, ent√£o agora podemos prosseguir diretamente para a an√°lise.

```python
# üîπ Verificar as primeiras linhas do dataset consolidado para garantir que os dados est√£o corretos
print("üîπ Primeiras linhas do dataset consolidado:")
display(df_completo.head())

# üîπ Exibir estat√≠sticas descritivas do dataset consolidado
print("\nüîπ Estat√≠sticas descritivas dos dados combinados:")
display(df_completo.describe())

# üîπ Exibir a quantidade de pa√≠ses analisados
print("\nüîπ Quantidade total de pa√≠ses analisados:", df_completo["Country"].nunique())
```

Agora vamos criar tr√™s gr√°ficos diferentes para analisar a evolu√ß√£o dos √≠ndices de felicidade.

###1Ô∏è‚É£ Gr√°fico de Barras ‚Äì Top 10 pa√≠ses mais felizes em 2019, 2020, 2022.

```python

# Filtrando os 10 pa√≠ses mais felizes de cada ano
top_10_2019 = df_completo[df_completo["Year"] == 2019].nlargest(10, "Happiness Score")
top_10_2020 = df_completo[df_completo["Year"] == 2020].nlargest(10, "Happiness Score")
top_10_2022 = df_completo[df_completo["Year"] == 2022].nlargest(10, "Happiness Score")

# Concatenando os tr√™s anos em um √∫nico DataFrame
df_top_10 = pd.concat([top_10_2019, top_10_2020, top_10_2022])

# Criando gr√°fico de barras agrupado
plt.figure(figsize=(12, 6))
sns.barplot(data=df_top_10, x="Country", y="Happiness Score", hue="Year", palette="viridis")

# Adicionando t√≠tulos e r√≥tulos
plt.title("üìä Compara√ß√£o dos 10 Pa√≠ses Mais Felizes (2019, 2020, 2022)", fontsize=14)
plt.xlabel("Pa√≠ses")
plt.ylabel("√çndice de Felicidade")
plt.xticks(rotation=45)  # Inclinar os nomes dos pa√≠ses para melhor leitura
plt.legend(title="Ano")  # Legenda indicando cada ano

# Exibindo gr√°fico
plt.show()
```

##2Ô∏è‚É£ Gr√°fico de Linha ‚Äì Evolu√ß√£o da Felicidade de Alguns Pa√≠ses de 2019, 2020, 2022.

```python
# Selecionar pa√≠ses espec√≠ficos para an√°lise da evolu√ß√£o da felicidade
paises_selecionados = ["Finland", "Denmark", "Norway", "Iceland", "Switzerland"]

# Filtrar o DataFrame apenas para esses pa√≠ses
df_paises = df_completo[df_completo["Country"].isin(paises_selecionados)]

# Criar o gr√°fico de linha
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_paises, x="Year", y="Happiness Score", hue="Country", marker="o", linewidth=2)

# Adicionar t√≠tulos e r√≥tulos
plt.title("üìà Evolu√ß√£o do √çndice de Felicidade (2019-2022)", fontsize=14)
plt.xlabel("Ano")
plt.ylabel("√çndice de Felicidade")
plt.xticks([2019, 2020, 2022])  # Garantir que apenas os anos analisados apare√ßam
plt.legend(title="Pa√≠ses")

# Exibir gr√°fico
plt.show()
```

##3Ô∏è‚É£ Gr√°fico de Dispers√£o ‚Äì PIB per Capita x Felicidade

```python
# Criando o gr√°fico de dispers√£o (scatter plot)
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_completo, x="GDP per capita", y="Happiness Score", hue="Year", palette="coolwarm", alpha=0.7)

# Adicionando t√≠tulos e r√≥tulos
plt.title("üí∞ Rela√ß√£o entre PIB per Capita e Felicidade", fontsize=14)
plt.xlabel("PIB per Capita")
plt.ylabel("√çndice de Felicidade")

# Exibindo gr√°fico
plt.show()
```
